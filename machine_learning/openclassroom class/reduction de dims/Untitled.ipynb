```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from sklearn import preprocessing, decomposition

%matplotlib inline
```

```python
data = pd.read_csv("wine.csv", header=None)
print(data.head())
print(data.describe())
```

## Setup Model

```python
X = data.as_matrix()
std_scale = preprocessing.StandardScaler()
X = std_scale.fit_transform(X)
```

```python
Sigma = np.dot(np.transpose(X), X)
print(Sigma.shape)
```

```python
eigenvalues, eignevectors = np.linalg.eigh(Sigma)
print(eigenvalues)
```

```python
plt.plot(eigenvalues, label = "Eigenvalues")
plt.legend()
plt.show()
```

```python
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eignevectors = eignevectors[:,idx]
```

```python
plt.plot(eigenvalues, label = "Eigenvalues")
plt.legend()
plt.show()
```

```python
t = np.sum(eigenvalues)
x = list(range(13))
y = []
for i in x:
    s = np.sum(eigenvalues[:i])
    print("Dim {} : {}".format(i+1, s/t))
    y.append(s/t)

plt.plot(x, y)
plt.show()
```

```python
print(eignevectors)
main_dir = eignevectors[:, :4]
print(main_dir)
```

```python
newX = np.dot(X, main_dir)
print(newX.shape)
```

## Verification

```python
X = data.as_matrix()
std_scale = preprocessing.StandardScaler()
X = std_scale.fit_transform(X)

pca = decomposition.PCA(n_components=4)
pca.fit_transform(X)

plt.scatter(X[:, 0], X[:, 1], c=X[:, 2], s=X[:, 3]*72)
plt.show()
```

```python
pcs = pca.components_

for i, (x, y) in enumerate(zip(pcs[0, :], pcs[1, :])):
    # Afficher un segment de l'origine au point (x, y)
    plt.plot([0, x], [0, y], color='k')
    # Afficher le nom (data.columns[i]) de la performance
    plt.text(x, y, data.columns[i], fontsize='14')

# Afficher une ligne horizontale y=0
plt.plot([-0.7, 0.7], [0, 0], color='grey', ls='--')

# Afficher une ligne verticale x=0
plt.plot([0, 0], [-0.7, 0.7], color='grey', ls='--')

plt.xlim([-0.7, 0.7])
plt.ylim([-0.7, 0.7])
```
